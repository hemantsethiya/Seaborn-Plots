## Seaborn  --

Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables. Visualization can be a core component of this process because, when data are visualized properly, the human visual system can see trends and patterns that indicate a relationship.

```python
import pandas as pd
import numpy as np
import seaborn as sns

data = sns.load_dataset('tips')
data.head()
```

###  Correlation with  Heatmap   --

```python
sns.heatmap(data.corr())
```

###  Jointplot   --

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = data, kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = data, kind = 'reg')
```

### Pairplot   --

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue = 'sex')
```

###  Dist plot

#####  it helps us to create histogram

```python
sns.distplot(data.tip)
```

###  Countplot --

```python
sns.countplot('sex', data = data)
```

```python
sns.countplot(y = 'sex', data = data)
```

###  Barplot   --

```python
sns.barplot(x = 'sex', y = 'total_bill', data=data)
```

###  Boxplot  --

```python
sns.boxplot('sex', 'total_bill', data = data)
```

```python
sns.boxplot('day','total_bill', data = data, palette = 'rainbow')
```

```python
sns.boxplot(data = data, orient = 'v')
```

```python
sns.boxplot('total_bill', 'day', hue = 'smoker', data = data)
```

###  Relplot  or scatter

```python
sns.relplot('total_bill','tip', data = data, hue = 'sex')
```

```python
sns.relplot('total_bill','tip', data = data, size = 'size', sizes = (25,200), hue = 'tip')
```

###  Catplot  (category)  --

```python
sns.catplot('sex', 'tip', data = data)
```

```python
sns.catplot('sex', 'tip', data = data, kind = 'swarm')
```

###  Violin Plot  --

```python
sns.violinplot(y="total_bill", x="day", data=data)
```
